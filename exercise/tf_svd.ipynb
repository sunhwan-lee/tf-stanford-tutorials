```python
import numpy as np
import tensorflow as tf
import time
```

```python
A = np.random.randn(9, 9)
print A
```

```python
# create placeholders for co-occurrence matrix
X = tf.placeholder(tf.float32, [9, 9], name='co_occurrence_matrix') 

# run SVD
s, u, v = tf.svd(X)

with tf.Session() as sess:

    print 'Running SVD...'
    
    start_time = time.time()

    s, u, v = sess.run([s, u, v], feed_dict={X: A})

    print 'Total time: {0} seconds'.format(time.time() - start_time)
    print 'SVD Finished!'
```

```python
print np.dot(u, np.dot(np.diag(s), v.T))
print A
```
